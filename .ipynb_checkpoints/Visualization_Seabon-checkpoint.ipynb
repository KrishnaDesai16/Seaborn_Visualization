```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb

```

```python
fifa_data=pd.read_csv('fifa.csv',index_col='Date',parse_dates=True)
```

```python
fifa_data.head()
```

```python
plt.figure(figsize=(12,6))
sb.lineplot(data=fifa_data)
```

```python
spotify=pd.read_csv('spotify.csv',parse_dates=True,index_col='Date')
```

```python
spotify.head()
```

```python
spotify.tail()
```

```python
spotify.fillna((spotify.mean()), inplace=True)
```

```python
spotify.head()
```

```python
plt.figure(figsize=(14,6))
plt.title('Daily global stream  of popular songs in 2017-2018')
sb.lineplot(data=spotify)
plt.show()
```

```python
#lineplot for subset of dataset
```

```python
sb.lineplot(data=spotify['Despacito'],label='Despacito')
sb.lineplot(data=spotify['Shape of You'],label='Shape of You')
plt.xlabel('Date')
plt.show()
```

```python
#barchart
flight_delay=pd.read_csv('flight_delays.csv',index_col='Month')
```

```python
flight_delay.head()
```

```python
plt.figure(figsize=(14,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')
sb.barplot(x=flight_delay.index,y=flight_delay.NK,data=flight_delay)
plt.show()
```

```python
#Heatmap
```

```python
plt.figure(figsize=(14,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')
sb.heatmap(data=flight_delay,annot=True)
plt.xlabel('Airlines Flight ')
plt.show()
```
