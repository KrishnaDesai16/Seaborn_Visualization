```python
spotify=pd.read_csv('spotify.csv',parse_dates=True,index_col='Date')
```

```python
spotify.head()
```

```python
#barchart
flight_delay=pd.read_csv('flight_delays.csv',index_col='Month')
```

```python
flight_delay.head()
```

```python
plt.figure(figsize=(14,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')
sb.barplot(x=flight_delay.index,y=flight_delay.NK,data=flight_delay)
plt.show()
```

```python
#Heatmap
```

```python
plt.figure(figsize=(14,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')
sb.heatmap(data=flight_delay,annot=True)
plt.xlabel('Airlines Flight ')
plt.show()
```

```python
data=pd.read_csv('insurance.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
sb.scatterplot(x=data['bmi'],y=data['charges'])
#it shows positively correaltion 
#As bmi increase -->insurrence charges will increase
```

```python
sb.regplot(x=data['bmi'],y=data['charges'])
```

```python
sb.scatterplot(x=data['bmi'],y=data['charges'],hue=data['smoker'])
#nonsmoker tend to pay slighlty as increasing in bmi wherease smoker  tend to pay much by increasing in bmi
```

```python
sb.lmplot(x='bmi',y='charges',data=data,hue='smoker')
#you can see that smokers has much steeper regression line relatively to nonsmokers line
```

```python
#categorical scatterplot
```

```python
sb.swarmplot(x=data['smoker'],
              y=data['charges'])
```

```python
a=pd.read_csv('setosa.csv',index_col='Id')
b=pd.read_csv('versicolor.csv',index_col='Id')
c=pd.read_csv('virginica.csv',index_col='Id')
```

```python
sb.distplot(a=a['Petal Length (cm)'],kde=False,label='setosa')
sb.distplot(a=b['Petal Length (cm)'],kde=False,label='versicolor')
sb.distplot(a=c['Petal Length (cm)'],kde=False,label='virginica')
plt.legend(loc=5)
```

```python
#kde plot
```

```python
sb.kdeplot(data=a['Petal Length (cm)'],shade=True,label='setosa')
sb.kdeplot(data=b['Petal Length (cm)'],shade=True,label='versicolor')
sb.kdeplot(data=c['Petal Length (cm)'],shade=True,label='virginica')
plt.legend(loc=0)
sb.set_style('whitegrid')
```
